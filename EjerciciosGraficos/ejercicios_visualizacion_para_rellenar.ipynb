## 1. Barplot: Distribuci√≥n de calificaciones

**Dataset:** [Students Performance in Exams (Kaggle)](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams)

**Pasos:**
1. Descarga el archivo CSV y c√°rgalo con pandas.
2. Realiza un barplot que muestre **el n√∫mero de estudiantes por grupo √©tnico** (`race/ethnicity`).

**Preguntas:**
- ¬øQu√© grupo es el m√°s numeroso?
- ¬øHay grupos con representaci√≥n especialmente baja?


```python
# Tu c√≥digo aqu√≠

!pip install seaborn 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("StudentsPerformance.csv")

# Mostrar conteo por grupo √©tnico
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="race/ethnicity", order=df["race/ethnicity"].value_counts().index, palette="pastel")
plt.title("N√∫mero de estudiantes por grupo √©tnico")
plt.xlabel("Grupo √©tnico")
plt.ylabel("Cantidad de estudiantes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. Boxplot: An√°lisis de salarios

## 9. Pairplot: Relaci√≥n entre atributos de flores

**Dataset:** Palmer Penguins (Seaborn)

**Pasos:**
1. Carga el dataset `penguins` usando seaborn.
2. Haz un pairplot de las variables num√©ricas, coloreando por especie.

**Preguntas:**
- ¬øAlguna combinaci√≥n de variables separa claramente las especies?
- ¬øAlguna variable parece poco √∫til para distinguir especies?


```python
# Tu c√≥digo aqu√≠
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset tips
tips = sns.load_dataset("tips")

# Preparar los datos para el gr√°fico apilado
stack_data = tips.groupby(['day', 'sex'])['total_bill'].sum().unstack()

# Gr√°fico de barras apiladas
plt.figure(figsize=(10, 6))
stack_data.plot(kind='bar', stacked=True, color=['lightcoral', 'skyblue'])
plt.title("Total de Facturaci√≥n por D√≠a y G√©nero")
plt.xlabel("D√≠a de la semana")
plt.ylabel("Total facturado ($)")
plt.legend(title="G√©nero")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
