## 1. Barplot: Distribución de calificaciones

**Dataset:** [Students Performance in Exams (Kaggle)](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams)

**Pasos:**
1. Descarga el archivo CSV y cárgalo con pandas.
2. Realiza un barplot que muestre **el número de estudiantes por grupo étnico** (`race/ethnicity`).

**Preguntas:**
- ¿Qué grupo es el más numeroso?
- ¿Hay grupos con representación especialmente baja?


```python
# Tu código aquí

!pip install seaborn 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("StudentsPerformance.csv")

# Mostrar conteo por grupo étnico
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="race/ethnicity", order=df["race/ethnicity"].value_counts().index, palette="pastel")
plt.title("Número de estudiantes por grupo étnico")
plt.xlabel("Grupo étnico")
plt.ylabel("Cantidad de estudiantes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Tu código aquí
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset mpg
mpg = sns.load_dataset("mpg")

# Calcular matriz de correlación
corr = mpg.corr(numeric_only=True)

# Crear heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")
plt.title("Matriz de Correlación entre Variables Numéricas")
plt.tight_layout()
plt.show()
```

## 9. Pairplot: Relación entre atributos de flores

**Dataset:** Palmer Penguins (Seaborn)

**Pasos:**
1. Carga el dataset `penguins` usando seaborn.
2. Haz un pairplot de las variables numéricas, coloreando por especie.

**Preguntas:**
- ¿Alguna combinación de variables separa claramente las especies?
- ¿Alguna variable parece poco útil para distinguir especies?


```python
# Tu código aquí
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset tips
tips = sns.load_dataset("tips")

# Preparar los datos para el gráfico apilado
stack_data = tips.groupby(['day', 'sex'])['total_bill'].sum().unstack()

# Gráfico de barras apiladas
plt.figure(figsize=(10, 6))
stack_data.plot(kind='bar', stacked=True, color=['lightcoral', 'skyblue'])
plt.title("Total de Facturación por Día y Género")
plt.xlabel("Día de la semana")
plt.ylabel("Total facturado ($)")
plt.legend(title="Género")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Tu código aquí
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar dataset diamonds
diamonds = sns.load_dataset("diamonds")

# Filtrar datos para mejor visualización
diamonds_sample = diamonds.sample(1000, random_state=42)

# Crear FacetGrid
g = sns.FacetGrid(diamonds_sample, col="cut", row="color", hue="clarity", 
                  palette="viridis", height=2.5, aspect=1.2)
g.map(sns.scatterplot, "carat", "price", alpha=0.7)
g.add_legend(title="Claridad")
g.fig.suptitle("Relación Quilate-Precio por Corte, Color y Claridad", y=1.05)
plt.tight_layout()
plt.show()
```
